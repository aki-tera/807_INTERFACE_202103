```python
# ハイパスフィルタの周波数応答を描画するプログラム

import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
def plot_freqz(b, a, title="0.5Hz 1st butterworth HighPassFilter"):
    
    w,h = signal.freqz(b, a, worN=np.logspace(-2,2,512), fs=Fs)
    
    plt.figure(figsize=[8,8])
    
    plt.subplot(211)
    plt.semilogx(w, 20*np.log10(np.abs(h)))
    plt.ylim(-45,5)
    plt.ylabel("Magnitude [dB]")
    plt.title(title)
    plt.grid(which="both", axis="both")
    
    plt.subplot(212)
    plt.semilogx(w, np.angle(h))
    plt.ylabel("Phase [rad]")
    plt.xlabel("Frequency [Hz]")
    plt.grid(which="both", axis="both")
    plt.suptitle("Frequency response")
    plt.show()
```

```python
Fs = 200             # サンプリング周波数
                     # 0.005sec = 5msec

Fh = 0.5             # ハイパス フィルタ遮断周波数
Nf = 1               # ハイパスフィルタ／ローパスフィルタの次数


# バターワース型フィルタの係数算出
# 21ページ目
bh, ah = signal.butter(Nf, Fh, "high", fs=Fs)
bl, al = signal.butter(Nf, Fl, "low", fs=Fs)


# ノッチフィルタの設計
# 21ページ目
Fn = 50.0            # ノッチフィルタ中心周波数
Q  = 4.0             # ノッチフィルタのQ値
bn, an = signal.iirnotch(Fn, Q, fs=Fs)
```

```python
# 一次バターワース型ハイパスフィルタの周波数応答
# 22ページ目

plot_freqz(bh, ah)
```
